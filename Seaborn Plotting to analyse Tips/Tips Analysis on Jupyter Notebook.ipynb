```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
# 1stº Plot - Distribution
sns.histplot(tips['total_bill'], bins = 30)

# argument bins is optional
# argument kde is optional
```

```python
# Jointplot -> Dispersion Plot

sns.jointplot(x='total_bill', y='tip', data= tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data= tips, kind ='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data= tips, kind ='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data= tips, kind ='kde')
```

```python
#PairPlot

sns.pairplot(tips)
```

```python
# we can analyse for a determined categorical variable, for example Sex; we need to add the argument 'hue' 

sns.pairplot(tips, hue='sex')
```

```python
# we can use a 3rd argument : pallette

sns.pairplot(tips, hue='sex', palette='coolwarm' )
```

```python
sns.pairplot(tips, hue='sex', palette ='magma')
```
